```python
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow as tf

# Define the custom model-building function
def build_vit_model(input_shape=(128, 128, 3), num_classes=2, patch_size=16, projection_dim=64, num_heads=4):
    inputs = keras.Input(shape=input_shape)

    # Patch embedding layer
    patches = layers.Conv2D(projection_dim, kernel_size=patch_size, strides=patch_size, padding='valid')(inputs)
    patches = layers.Reshape((-1, projection_dim))(patches)

    # Positional encoding
    position_embedding = layers.Embedding(input_dim=patches.shape[1], output_dim=projection_dim)(
        tf.range(start=0, limit=patches.shape[1])
    )
    x = patches + position_embedding

    # Transformer Encoder Blocks
    for _ in range(4):
        x1 = layers.LayerNormalization()(x)
        attention_output = layers.MultiHeadAttention(num_heads=num_heads, key_dim=projection_dim)(x1, x1)
        x2 = layers.Add()([x, attention_output])
        x3 = layers.LayerNormalization()(x2)
        x3 = layers.Dense(projection_dim, activation='relu')(x3)
        x = layers.Add()([x2, x3])

    # Classification Head
    x = layers.LayerNormalization()(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(64, activation='relu')(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = keras.Model(inputs, outputs)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model


# Rebuild the model architecture
model = build_vit_model()

# Load the trained weights
model.load_weights('vegetation_vit_model.h5')
print("Model weights loaded successfully.")
```

```python
import numpy as np
from PIL import Image
import os

# Define paths
test_path = './data/test'
segmented_images_path = './segmented_images'

# Create the segmented images directory if it doesn't exist
if not os.path.exists(segmented_images_path):
    os.makedirs(segmented_images_path)


# Function to predict and save segmented image
def segment_and_save_image(image_path, model, save_path):
    # Load and preprocess image
    image = Image.open(image_path).resize((128, 128))
    image_array = np.array(image) / 255.0
    image_array = np.expand_dims(image_array, axis=0)

    # Make prediction
    prediction = model.predict(image_array)
    predicted_class = np.argmax(prediction, axis=1)[0]

    # Save the segmented image
    segmented_image_path = os.path.join(save_path, os.path.basename(image_path))
    image.save(segmented_image_path)
    print(f"Segmented image saved at: {segmented_image_path}")


# Test the model on all images in the test directory
test_images = os.listdir(test_path)
for img_name in test_images:
    img_path = os.path.join(test_path, img_name)
    segment_and_save_image(img_path, model, segmented_images_path)
```
